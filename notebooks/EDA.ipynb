# **EDA**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("yield_df.csv")


print("Data Awal:")
print(df.head())

print("\n Informasi Dataset:")
print(df.info())

# Statistik Deskriptif
print("\n Statistik Deskriptif:")
print(df.describe())

print("\n Nilai Unik:")
print("Area:", df['Area'].unique())
print("Komoditas:", df['Yield_Value'].unique())
print("Tahun:", df['Year'].unique())

print("\n Korelasi antar Fitur Numerik:")
print(df.corr(numeric_only=True))

print("\nMissing Values:")
print(df.isnull().sum())
```

# **Visualisasi**

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Korelasi Antar Variabel")
plt.tight_layout()
plt.show()


plt.subplots_adjust(hspace=0.5)

plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x="Year", y="Yield_Value", marker='o')
plt.title(" Tren Yield per Tahun")
plt.ylabel("Yield (kg/ha)")
plt.grid(True)
plt.tight_layout()
plt.show()

plt.subplots_adjust(hspace=0.5)

plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x="Pesticide_Value", y="Yield_Value", hue="Year", palette="viridis")
plt.title(" Yield vs Pesticide Use")
plt.xlabel("Penggunaan Pestisida")
plt.ylabel("Yield")
plt.tight_layout()
plt.show()

plt.subplots_adjust(hspace=0.5)

plt.figure(figsize=(8, 5))
sns.histplot(df["Avg_Temperature_celsius"], bins=10, kde=True)
plt.title(" Distribusi Suhu Rata-rata")
plt.tight_layout()
plt.show()

plt.subplots_adjust(hspace=0.5)

plt.figure(figsize=(8, 5))
sns.histplot(df["Avg_Rainfall_mm"], bins=10, kde=True, color='skyblue')
plt.title(" Distribusi Curah Hujan Rata-rata")
plt.tight_layout()
plt.show()
```
